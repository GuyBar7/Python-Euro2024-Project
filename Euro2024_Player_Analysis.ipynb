#### <span style="color:green">Question 3: Top 5 Players with the Highest Goals per Cap</span>
In this section, we identify the top 5 players with the highest goals per cap by calculating the ratio of goals to caps and filtering for those with more than 10 goals and 10 caps.

````python
# Calculate the Goals per Cap ratio
df['Goals_per_Cap'] = df['Goals'] / df['Caps']

# Handle cases where Caps is 0 to avoid division by zero
df['Goals_per_Cap'] = df['Goals_per_Cap'].fillna(0)

# Filter players with more than 10 goals and more than 10 caps
filtered_df = df[(df['Goals'] > 10) & (df['Caps'] > 10)]

# Sort the filtered DataFrame by Goals_per_Cap in descending order and select the top 5 players
top_players_by_goals_per_cap = filtered_df.sort_values(by='Goals_per_Cap', ascending=False).head(5)

# Reset index to have a clean numerical index
top_players_by_goals_per_cap.reset_index(drop=True, inplace=True)

# Add 1 to the index for display purposes
top_players_by_goals_per_cap.index += 1

# Display the top 5 players with their name, goals, caps, goals per cap ratio, and country
print("Top 5 players with the highest goals per cap (more than 10 goals and 10 caps):")
print(top_players_by_goals_per_cap[['Name', 'Country', 'Goals', 'Caps', 'Goals_per_Cap']])
````

<span style="color:red; text-decoration:underline;">Conclusion:</span>
The data shows that Romelu Lukaku has the highest goals per cap ratio (0.728), highlighting his impressive efficiency in front of the goal. Harry Kane is second with a ratio of 0.692, demonstrating his consistent scoring ability for England. Niclas Füllkrug, Aleksandar Mitrovic, and Cristiano Ronaldo round out the top five, showcasing their strong goal-scoring records for Germany, Serbia, and Portugal, respectively. These players significantly contribute to their national teams with their high scoring rates.
Measuring the goals per cap ratio is important as it provides insight into a player's effectiveness and impact on the field, allowing teams and analysts to assess their value beyond just the total number of goals scored.

#### <span style="color:green">Question 4: Scatter Plot of Caps vs. Goals</span>
We create a scatter plot to visualize the relationship between the number of caps and the number of goals scored by players. This helps to understand if there is a correlation between these two metrics.

````python
# Create a custom palette for unique colors
palette = sns.color_palette('husl', len(df['Position'].unique()))

# Scatter plot of Caps vs. Goals with unique colors for each position
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Caps', y='Goals', data=df, hue='Position', palette=palette, alpha=0.7, s=200)
plt.title('Relationship Between Caps and Goals')
plt.xlabel('Number of Caps (Appearances)', fontsize=14)
plt.ylabel('Number of Goals', fontsize=14)
plt.legend(title='Position', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
````

<span style="color:red; text-decoration:underline;">Conclusion:</span>
The scatter plot visualizes the relationship between the number of caps (appearances) and the number of goals across different player positions. The plot shows a general trend where players with more caps tend to have more goals, though this varies significantly by position. For example, attacking positions such as Centre-Forwards and Second Strikers (shown in purple and blue) generally have higher goals per cap, reflecting their role as primary goal scorers. In contrast, defensive players like Goalkeepers and Centre-Backs have fewer goals regardless of their number of appearances, as scoring is not their primary function.

An example from the plot is Cristiano Ronaldo, represented by one of the high points in the upper right quadrant. As a Centre-Forward, Ronaldo has accumulated a significant number of goals due to his long career and consistent goal-scoring ability, highlighting the expected trend for players in forward positions.

#### <span style="color:green">Question 11: Distribution of Player Ages</span>
We create a histogram to visualize the distribution of player ages. This helps to understand the age range and concentration of players within different age groups.

````python
# Calculate statistics for player ages
average_age = df['Age'].mean()
std_dev_age = df['Age'].std()
median_age = df['Age'].median()

# Distribution of Player Ages
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=np.arange(df['Age'].min(), df['Age'].max() + 1), kde=True, color='blue')
plt.title('Distribution of Player Ages')
plt.xlabel('Age', fontsize=14)
plt.ylabel('Number of Players', fontsize=14)
plt.xticks(np.arange(df['Age'].min(), df['Age'].max() + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add text for statistics
plt.text(df['Age'].max() - 2, plt.gca().get_ylim()[1] * 0.8, f'Average Age: {average_age:.2f}', 
         fontsize=12, fontweight='bold', color='black', ha='right')

plt.text(df['Age'].max() - 2, plt.gca().get_ylim()[1] * 0.75, f'Std Dev: {std_dev_age:.2f}', 
         fontsize=12, fontweight='bold', color='black', ha='right')

plt.text(df['Age'].max() - 2, plt.gca().get_ylim()[1] * 0.7, f'Median Age: {median_age}', 
         fontsize=12, fontweight='bold', color='black', ha='right')

plt.show()
````

<span style="color:red; text-decoration:underline;">Conclusion:</span>
The histogram shows the distribution of player ages, with most players clustered around the average age of 27.04 years. The distribution is approximately normal, with a slight skew towards younger players. The standard deviation of 4.12 indicates a moderate spread in ages, while the median age of 27.0 aligns closely with the mean, suggesting a balanced age distribution. Analyzing age distribution helps teams understand the overall maturity and experience level of players, which can influence team dynamics, strategy, and planning for future player development or transfers.

#### <span style="color:green">Question 12: Top 10 Clubs by Market Value</span>
We find the top 10 clubs with the highest market value. This is done by sorting the data based on the 'MarketValue' column and selecting the top 10 clubs.

````python
# Calculating total market value for each club and selecting top 10
df_group = df.groupby('Club')['MarketValue'].sum()
top_10_clubs = df_group.sort_values(ascending=False).head(10)

# Plotting the bar chart using seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_clubs.values, y=top_10_clubs.index, hue=top_10_clubs.index, dodge=False, palette='Set1', legend=False)
plt.title('Top 10 Clubs by Market Value')

# Adding values and club names inside the bars with formatted numbers
for i, (club, value) in enumerate(top_10_clubs.items()):
    formatted_value = f'{value / 1e6:.0f}M €'  # Format the number in millions with Euro symbol
    plt.text(value / 2, i, f'{club}\n{formatted_value}', ha='center', va='center', color='black', fontsize=10, fontweight='bold')

# Hiding x and y axis labels and titles
plt.xlabel('')  # Remove x-axis label
plt.ylabel('')  # Remove y-axis label
plt.xticks([])  # Hide x-axis ticks
plt.yticks([])  # Hide y-axis ticks

plt.grid(False)  # Hide grid lines

# Display the plot
plt.show()
````

<span style="color:red; text-decoration:underline;">Conclusion:</span>
The chart shows the top 10 clubs by market value, with Manchester City leading at €826 million, followed by Paris Saint-Germain and Arsenal FC. This highlights the financial power of these clubs, which can impact their ability to sign top players and succeed in competitions. In the context of Euro 2024, many players from these high-value clubs are likely to participate, suggesting that clubs with higher market values might have more players represented in the tournament, potentially influencing their national teams' performance.

#### <span style="color:green">Question 17: England vs. Spain Comparison</span>
This question compares player statistics for England and Spain, including metrics like average age, market value, goals, and caps.

````python
def calculate_statistics(df):
    # Calculate required statistics
    stats = {
        'Number of Players': len(df),
        'Average Age': df['Age'].mean(),
        'Std Dev Age': df['Age'].std(),
        'Average Market Value': df['MarketValue'].mean() / 1e6,  # Convert to millions
        'Average Goals': df['Goals'].mean(),
        'Total Goals': df['Goals'].sum(),
        'Average Caps': df['Caps'].mean(),
        'Total Caps': df['Caps'].sum(),
        'Average Height': df['Height'].mean(),
        'Youngest Player': f"{df.loc[df['Age'].idxmin()]['Name']} ({df['Age'].min()} years)",
        'Oldest Player': f"{df.loc[df['Age'].idxmax()]['Name']} ({df['Age'].max()} years)",
        'Highest Market Value Player': f"{df.loc[df['MarketValue'].idxmax()]['Name']} (${df['MarketValue'].max() / 1e6:.1f} M)",
        'Top Scorer': f"{df.loc[df['Goals'].idxmax()]['Name']} ({df['Goals'].max()} goals)",
        'Most Caps Player': f"{df.loc[df['Caps'].idxmax()]['Name']} ({df['Caps'].max()} caps)",
        'Tallest Player': f"{df.loc[df['Height'].idxmax()]['Name']} ({df['Height'].max()} cm)"
    }
    return stats

# Filter data for Spain and England
data = {
    'Spain': calculate_statistics(df[df['Country'] == 'Spain']),
    'England': calculate_statistics(df[df['Country'] == 'England'])
}

# Create a DataFrame for the comparison with the requested order
metrics = [
    'Number of Players', 
    'Average Age', 
    'Std Dev Age',
    'Average Market Value', 
    'Average Goals', 
    'Total Goals', 
    'Average Caps', 
    'Total Caps',
    'Average Height',
    'Youngest Player',
    'Oldest Player',
    'Highest Market Value Player',
    'Top Scorer',
    'Most Caps Player',
    'Tallest Player'
]

comparison_df = pd.DataFrame({
    'Metric': metrics,
    'Spain': [data['Spain'][metric] for metric in metrics],
    'England': [data['England'][metric] for metric in metrics]
})

# Apply formatting to the DataFrame
comparison_df['Spain'] = comparison_df.apply(lambda row: format_value(row['Spain'], row['Metric']), axis=1)
comparison_df['England'] = comparison_df.apply(lambda row: format_value(row['England'], row['Metric']), axis=1)

# Calculate the length of the longest string for formatting the separator line
max_length = comparison_df[['Metric', 'Spain', 'England']].apply(lambda x: x.astype(str).str.len()).max().max()
separator_line = '-' * (max_length + 36)  # Adjust width as needed for proper separation

# Print formatted table with a separator line
print("\nComparison of Spain vs England:\n")
print(f"{'Metric':<{max_length}} {'Spain':<30} {'England':<30}")
print(separator_line)
for index, row in comparison_df.iterrows():
    print(f"{row['Metric']:<{max_length}} {row['Spain']:<30} {row['England']:<30}")
````

#### <span style="color:green">Question 18: Analysis of Euro Winners and Years Since Last Victory</span>
This question shows the Euro-winning teams, the years they won, and how many years have passed since their last victory.

````python
# List of Euro winners and their winning years
euro_winners = {
    'Spain': [1964, 2008, 2012, 2024],
    'Italy': [1968, 2021],
    'Portugal': [2016],
    'France': [1984, 2000],
    'Germany': [1972, 1980, 1996],
    'Greece': [2004],
    'Netherlands': [1988],
    'Denmark': [1992],
    'Czechoslovakia': [1976],
    'Soviet Union': [1960]
}

# Create a DataFrame for the Euro winners
df_euro_winners = pd.DataFrame(list(euro_winners.items()), columns=['Country', 'Winning Years'])

# Function to calculate years since last Euro win
def years_since_last_win(years):
    current_year = datetime.now().year
    last_win = max(years)  # Get the most recent win
    return current_year - last_win

# Add a new column for the years since last Euro win
df_euro_winners['Years Since Last Win'] = df_euro_winners['Winning Years'].apply(years_since_last_win)

# Merge the Euro winners data with your existing dataset
df_merged = df.merge(df_euro_winners, how='left', on='Country')

# Convert 'Winning Years' from list to string for better processing and display
df_merged['Winning Years'] = df_merged['Winning Years'].apply(lambda x: ', '.join(map(str, x)) if isinstance(x, list) else x)

# Remove rows where 'Winning Years' is NaN and create a copy to avoid warnings
df_winners_only = df_merged.dropna(subset=['Winning Years']).copy()

# Update 'Years Since Last Win' using .loc to avoid SettingWithCopyWarning
df_winners_only.loc[:, 'Years Since Last Win'] = df_winners_only['Winning Years'].apply(lambda x: years_since_last_win(list(map(int, x.split(', ')))) if isinstance(x, str) else None)

# Sort the DataFrame by 'Years Since Last Win' in ascending order and reset the index
df_sorted = df_winners_only[['Country', 'Winning Years', 'Years Since Last Win']].drop_duplicates().sort_values(by='Years Since Last Win', ascending=True).reset_index(drop=True)

# Display the sorted DataFrame without index
print(df_sorted.to_string(index=False))
````
